```python
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
matplotlib.use('TkAgg')

##---------------------------------------------------
## Step 0: Install all other dependencies that occur at run time if  any module not found.
##---------------------------------------------------
print("done")
```

```python
Number_of_points = 20   # recommended >=20
img = cv2.imread("Indian_Flag.jpg")

def select_points(img, title):
    fig, ax = plt.subplots()
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    ax.imshow(img_rgb)
    ax.set_title(title)

    fig.canvas.manager.set_window_title('Select Points')

    cursor = matplotlib.widgets.Cursor(ax, useblit=True, color='red', linewidth=1)

    plt.show(block=False)

    k = 0
    points = []

    while k < Number_of_points:
        xy = plt.ginput(1, timeout=0)
        if len(xy) > 0:
            col, row = map(int, xy[0])

            # Step 3: Collect RGB values
            rgb_value = img_rgb[row, col]
            print(f"RGB at ({row},{col}):", rgb_value)

            points.append([row, col, rgb_value])

            plt.scatter(col, row, c='black', marker='o', s=10)
            plt.draw()

            k += 1

    plt.close()
    return points
```

```python
pts_saffron = select_points(img, "Select Saffron Strip Points")
pts_white   = select_points(img, "Select White Strip Points")
pts_green   = select_points(img, "Select Green Strip Points")
```

```python
def rgb_to_lab(rgb):
    return cv2.cvtColor(np.uint8([[rgb]]), cv2.COLOR_RGB2Lab)[0][0]

saffron_lab = np.array([rgb_to_lab(rgb) for _, _, rgb in pts_saffron])
white_lab   = np.array([rgb_to_lab(rgb) for _, _, rgb in pts_white])
green_lab   = np.array([rgb_to_lab(rgb) for _, _, rgb in pts_green])

# Extract a* and b*
a_features = np.hstack((saffron_lab[:,1], white_lab[:,1], green_lab[:,1]))
b_features = np.hstack((saffron_lab[:,2], white_lab[:,2], green_lab[:,2]))
```

```python
class_mapping = {'Saffron':0, 'White':1, 'Green':2}

y = np.array(
    [0]*Number_of_points +
    [1]*Number_of_points +
    [2]*Number_of_points
)

plt.figure()

plt.scatter(a_features[:Number_of_points],
            b_features[:Number_of_points],
            marker='o', label='Saffron')

plt.scatter(a_features[Number_of_points:2*Number_of_points],
            b_features[Number_of_points:2*Number_of_points],
            marker='^', label='White')

plt.scatter(a_features[2*Number_of_points:],
            b_features[2*Number_of_points:],
            marker='*', label='Green')

plt.xlabel('Feature A (a*)')
plt.ylabel('Feature B (b*)')
plt.title('Color Distribution')
plt.legend()
plt.grid()
plt.show()

lda = LDA()
lda.fit(np.column_stack((a_features, b_features)), y)
```

```python
plt.figure()

plt.scatter(a_features[:Number_of_points],
            b_features[:Number_of_points], marker='o', label='Saffron')

plt.scatter(a_features[Number_of_points:2*Number_of_points],
            b_features[Number_of_points:2*Number_of_points], marker='^', label='White')

plt.scatter(a_features[2*Number_of_points:],
            b_features[2*Number_of_points:], marker='*', label='Green')

plt.xlabel('Feature A (a*)')
plt.ylabel('Feature B (b*)')
plt.title('LDA boundaries for Colors of the Indian Flag')

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(
    np.linspace(xlim[0], xlim[1], 100),
    np.linspace(ylim[0], ylim[1], 100)
)

Z = lda.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contour(xx, yy, Z, colors='k', linewidths=2)

plt.legend()
plt.grid()
plt.show()
```
